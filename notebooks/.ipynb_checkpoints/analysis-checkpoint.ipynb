```python
from matplotlib.lines import Line2D
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.cm import ScalarMappable

NaN = float('nan')

data = {
    'PTMSP': {'P_O2': 3000.0, 'P_N2': 1800.0, 'P_CO2': 19000.0, 'P_CH4': 4300.0, 'Ed_O2': 5.0, 'Ed_N2': 5.0, 'Ed_CO2': 0.4, 'Ed_CH4': 4.9},
    'Hyflon AD60x': {'P_O2': 121.0, 'P_N2': 49.0, 'P_CO2': 304.0, 'P_CH4': 33.0, 'Ed_O2': 3.65, 'Ed_N2': 4.64, 'Ed_CO2': 3.89, 'Ed_CH4': 3.9},
    'AD60x + F4 MIL140': {'P_O2': 108.0, 'P_N2': 38.0, 'P_CO2': 290.0, 'P_CH4': 20.0, 'Ed_O2': 4.25, 'Ed_N2': 5.19, 'Ed_CO2': 3.09, 'Ed_CH4': 7.12},
    'Matrimid': {'P_O2': 2.4, 'P_N2': 0.47, 'P_CO2': 15.04, 'P_CH4': 0.54, 'Ed_O2': 6.35, 'Ed_N2': 7.46, 'Ed_CO2': 6.81, 'Ed_CH4': 7.96},
    'Matrimid + Trip-H_20': {'P_O2': 5.28, 'P_N2': 0.97, 'P_CO2': 29.6, 'P_CH4': 0.92, 'Ed_O2': 7.91, 'Ed_N2': 8.41, 'Ed_CO2': 7.06, 'Ed_CH4': 10.22},
    'Matrimid + Trip-NH2_20': {'P_O2': 3.64, 'P_N2': 0.57, 'P_CO2': 20.3, 'P_CH4': 1.0, 'Ed_O2': 7.6, 'Ed_N2': 8.48, 'Ed_CO2': 7.78, 'Ed_CH4': 10.74},
    'Matrimid + Trip-NO2_20': {'P_O2': 4.86, 'P_N2': 1.1, 'P_CO2': 25.6, 'P_CH4': 1.0, 'Ed_O2': 6.2, 'Ed_N2': 7.46, 'Ed_CO2': 6.69, 'Ed_CH4': 9.38},
    'Matrimid + Trip-SO3H_20': {'P_O2': 2.75, 'P_N2': 0.54, 'P_CO2': 16.1, 'P_CH4': 0.49, 'Ed_O2': 6.49, 'Ed_N2': 7.22, 'Ed_CO2': 6.64, 'Ed_CH4': 9.36},
    'PIM-TOT-Cardo': {'P_O2': 186.0, 'P_N2': 46.0, 'P_CO2': 1026.0, 'P_CH4': 52.3, 'Ed_O2': 4.44, 'Ed_N2': 5.8, 'Ed_CO2': 4.17, 'Ed_CH4': 6.78},
    'PIM-TOT-SBF-5': {'P_O2': 1960.0, 'P_N2': 717.0, 'P_CO2': 11400.0, 'P_CH4': 1430.0, 'Ed_O2': 3.25, 'Ed_N2': 4.82, 'Ed_CO2': 3.42, 'Ed_CH4': 5.92},
    'PIM-SBI-Trip': {'P_O2': 6390.0, 'P_N2': 2315.0, 'P_CO2': 35600.0, 'P_CH4': 4600.0, 'Ed_O2': 7.23, 'Ed_N2': 12.33, 'Ed_CO2': 2.37, 'Ed_CH4': 2.1},
    'PIM-SBI-Trip-co_PIM-1': {'P_O2': 4230.0, 'P_N2': 1298.0, 'P_CO2': 24410.0, 'P_CH4': 2519.0, 'Ed_O2': 6.4, 'Ed_N2': 10.84, 'Ed_CO2': 1.8, 'Ed_CH4': 1.98},
    'PIM-1': {'P_O2': 1580.0, 'P_N2': 506.0, 'P_CO2': 9650.0, 'P_CH4': 913.0, 'Ed_O2': 4.2, 'Ed_N2': 7.8, 'Ed_CO2': 3.2, 'Ed_CH4': 3.0},
    'PIM-TMN-Trip': {'P_O2': 5431.1, 'P_N2': 1705.6, 'P_CO2': 25484.2, 'P_CH4': 2603.3, 'Ed_O2': 2.62, 'Ed_N2': 4.5, 'Ed_CO2': 3.01, 'Ed_CH4': 5.39},
    'PIM-BTrip': {'P_O2': 2368.5, 'P_N2': 505.0, 'P_CO2': 11292.2, 'P_CH4': 536.1, 'Ed_O2': 4.9, 'Ed_N2': 7.45, 'Ed_CO2': 5.4, 'Ed_CH4': 8.78},
    '6FDA-DAM': {'P_O2': 328.0, 'P_N2': 95.8, 'P_CO2': 1721.0, 'P_CH4': NaN, 'Ed_O2': 4.0, 'Ed_N2': 4.6, 'Ed_CO2': 3.3, 'Ed_CH4': NaN},
    '6FDA-PMDA-FDA': {'P_O2': 33.3, 'P_N2': 7.41, 'P_CO2': 192.0, 'P_CH4': NaN, 'Ed_O2': 5.07, 'Ed_N2': 6.18, 'Ed_CO2': 4.99, 'Ed_CH4': NaN},
    'NBPI': {'P_O2': 32.5, 'P_N2': 6.88, 'P_CO2': 171.0, 'P_CH4': 7.76, 'Ed_O2': 4.01, 'Ed_N2': 6.27, 'Ed_CO2': 4.76, 'Ed_CH4': 7.02},
    'DBPI': {'P_O2': 273.0, 'P_N2': 59.2, 'P_CO2': 1382.0, 'P_CH4': 73.4, 'Ed_O2': 4.13, 'Ed_N2': 7.20, 'Ed_CO2': 5.31, 'Ed_CH4': 9.85},
    'NBPI-550': {'P_O2': 536.0, 'P_N2': 96.7, 'P_CO2': 2514.0, 'P_CH4': 87.8, 'Ed_O2': 4.29, 'Ed_N2': 9.09, 'Ed_CO2': 5.92, 'Ed_CH4': 12.68},
    'DBPI-550': {'P_O2': 4274.0, 'P_N2': 708.0, 'P_CO2': 20639.0, 'P_CH4': 668.0, 'Ed_O2': 6.17, 'Ed_N2': 9.98, 'Ed_CO2': 6.30, 'Ed_CH4': 12.84},
    'PIM-PI-1': {'P_O2': 400.0, 'P_N2': 110.0, 'P_CO2': 1966.0, 'P_CH4': 157.0, 'Ed_O2': 5.35, 'Ed_N2': 6.79, 'Ed_CO2': 5.3, 'Ed_CH4': 8.84},
    'PIM-DB-PI': {'P_O2': 151.0, 'P_N2': 33.0, 'P_CO2': 764.0, 'P_CH4': 33.0, 'Ed_O2': 4.01, 'Ed_N2': 6.05, 'Ed_CO2': 4.23, 'Ed_CH4': 7.47},
    'Precursor membrane (6FDA/DETDA:DABA)': {'P_O2': 54, 'P_N2': 15.42, 'P_CO2': 238, 'P_CH4': 10.81, 'Ed_O2': 3.2505, 'Ed_N2': 3.7046, 'Ed_CO2': 3.2266, 'Ed_CH4': 3.9914},
    '550C-CMS (6FDA/DETDA:DABA)': {'P_O2': 4293.0, 'P_N2': 868.0, 'P_CO2': 21740.0, 'P_CH4': 723.0, 'Ed_O2': 3.7524, 'Ed_N2': 4.1826, 'Ed_CO2': 3.5373, 'Ed_CH4': 4.3977},
    '800C-CMS (6FDA/DETDA:DABA)': {'P_O2': 300, 'P_N2': 39.47, 'P_CO2': 1812, 'P_CH4': 20.52, 'Ed_O2': 5.8795, 'Ed_N2': 6.0946, 'Ed_CO2': 5.8078, 'Ed_CH4': 6.262},
    '675C-CMS (6FDA/DETDA:DABA)': {'P_O2': 928, 'P_N2': 123.7, 'P_CO2': 5569, 'P_CH4': 101.2, 'Ed_O2': 5.3776, 'Ed_N2': 5.521, 'Ed_CO2': 5.3298, 'Ed_CH4': 5.7361},
    'a_gTIF-4': {'P_O2': NaN, 'P_N2': NaN, 'P_CO2': 266.5, 'P_CH4': 13.18, 'Ed_O2': NaN, 'Ed_N2': NaN, 'Ed_CO2': 1.544, 'Ed_CH4': 3.7882},
    'a_gZIF-62': {'P_O2': NaN, 'P_N2': NaN, 'P_CO2': 313.2, 'P_CH4': 13.15, 'Ed_O2': NaN, 'Ed_N2': NaN, 'Ed_CO2': 2.8872, 'Ed_CH4': 5.2605},
}

upperbounds_dict = {
    "CO2_CH4": {"1991": [1073700, -2.63], "2008": [5369140, -2.64], "2019": [22584455, -2.40], "MIX2018": [6398467, -2.99]},
    "CO2_N2":  {"2008": [30967000, -2.89], "2019": [755584213, -3.41]},
    "O2_N2":   {"1991": [389224, -5.80], "2008": [1396000, -5.67], "2015": [16700000, -5.7]},
}

gas_pair_to_ub_key = {
    ('CO2', 'N2'):  'CO2_N2',
    ('CO2', 'CH4'): 'CO2_CH4',
    ('O2',  'N2'):  'O2_N2',
}

ub_line_styles = {
    '1991':    {'ls': ':',              'lw': 1.8, 'color': '#e67e22'},
    '2008':    {'ls': '--',             'lw': 1.8, 'color': '#c0392b'},
    '2015':    {'ls': '-.',             'lw': 1.8, 'color': '#8e44ad'},
    '2019':    {'ls': '-',              'lw': 1.8, 'color': '#2980b9'},
    'MIX2018': {'ls': (0,(3,1,1,1)),   'lw': 1.8, 'color': '#27ae60'},
}

plots_config = [
    ('CO2', 'N2',  'Ed_CO2', 'Ed_N2',  r'$\Delta E_\mathrm{d}$ (CO$_2$/N$_2$) (kJ mol$^{-1}$)'),
    ('CO2', 'CH4', 'Ed_CO2', 'Ed_CH4', r'$\Delta E_\mathrm{d}$ (CO$_2$/CH$_4$) (kJ mol$^{-1}$)'),
    ('O2',  'N2',  'Ed_O2',  'Ed_N2',  r'$\Delta E_\mathrm{d}$ (O$_2$/N$_2$) (kJ mol$^{-1}$)'),
]

plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.sans-serif': ['Liberation Sans', 'Arial', 'Helvetica', 'DejaVu Sans'],
    'axes.linewidth': 1.2,
    'xtick.direction': 'in', 'ytick.direction': 'in',
    'xtick.top': True, 'ytick.right': True,
    'xtick.major.size': 5, 'ytick.major.size': 5,
    'xtick.minor.size': 3, 'ytick.minor.size': 3,
    'xtick.minor.visible': True, 'ytick.minor.visible': True,
})

all_ed = []
for props in data.values():
    for _, _, edA_key, edB_key, _ in plots_config:
        edA = props.get(edA_key, NaN)
        edB = props.get(edB_key, NaN)
        if not np.isnan(edA) and not np.isnan(edB):
            all_ed.append(edA - edB)

abs_max = max(abs(min(all_ed)), abs(max(all_ed)))
norm = mcolors.TwoSlopeNorm(vmin=-abs_max, vcenter=0, vmax=abs_max)
cmap = plt.cm.coolwarm

fig, axes = plt.subplots(1, 3, figsize=(18, 5.5))
fig.patch.set_facecolor('white')

for ax, (gA, gB, edA_key, edB_key, cbar_label) in zip(axes, plots_config):
    ax.set_facecolor('white')

    p_key_A = f'P_{gA}'
    p_key_B = f'P_{gB}'

    P_A_vals, sel_vals, ed_vals, labels = [], [], [], []
    for mat, props in data.items():
        pA  = props.get(p_key_A, NaN)
        pB  = props.get(p_key_B, NaN)
        edA = props.get(edA_key, NaN)
        edB = props.get(edB_key, NaN)
        if any(np.isnan(x) for x in [pA, pB, edA, edB]) or pB == 0:
            continue
        P_A_vals.append(pA)
        sel_vals.append(pA / pB)
        ed_vals.append(edA - edB)
        labels.append(mat)

    if not P_A_vals:
        continue

    pad_factor = 0.45
    p_min = 10 ** (np.log10(min(P_A_vals)) - pad_factor)
    p_max = 10 ** (np.log10(max(P_A_vals)) + pad_factor)
    s_min = 10 ** (np.log10(min(sel_vals)) - pad_factor)
    s_max = 10 ** (np.log10(max(sel_vals)) + pad_factor)

    ub_key = gas_pair_to_ub_key.get((gA, gB))
    ub_handles = []
    if ub_key and ub_key in upperbounds_dict:
        y_range = np.logspace(np.log10(s_min * 0.5), np.log10(s_max * 2.0), 500)
        for year, (k, n) in sorted(upperbounds_dict[ub_key].items()):
            x_ub = k * y_range ** n
            mask = (x_ub >= p_min) & (x_ub <= p_max)
            if not mask.any():
                continue
            sty = ub_line_styles.get(year, {'ls': '--', 'lw': 1.5, 'color': 'gray'})
            ax.plot(x_ub[mask], y_range[mask], **sty, alpha=0.7, zorder=2)
            ub_handles.append(Line2D([0], [0], **sty, alpha=0.85, label=f'UB {year}'))

    sc = ax.scatter(
        P_A_vals, sel_vals,
        c=ed_vals, cmap=cmap, norm=norm,
        s=70, edgecolors='#333333', linewidths=0.6, zorder=5
    )

    for x, y, lbl in zip(P_A_vals, sel_vals, labels):
        ax.annotate(lbl, (x, y), fontsize=7.5, color='#1a1a1a',
                    xytext=(5, 4), textcoords='offset points')

    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlim(p_min, p_max)
    ax.set_ylim(s_min, s_max)

    ax.set_xlabel(fr'$P_{{\mathrm{{{gA}}}}}$ (Barrer)', fontsize=13)
    ax.set_ylabel(fr'$\alpha_{{\mathrm{{{gA}/{gB}}}}}$', fontsize=13)
    ax.set_title(fr'{gA}/{gB} Separation', fontsize=13, fontweight='bold', pad=8)
    ax.tick_params(axis='both', which='both', labelsize=11)

    if ub_handles:
        ax.legend(handles=ub_handles, fontsize=8.5, frameon=True,
                  framealpha=0.9, edgecolor='#aaaaaa',
                  facecolor='white', loc='upper left',
                  bbox_to_anchor=(0.01, 0.99))

    for spine in ax.spines.values():
        spine.set_linewidth(1.2)
        spine.set_color('#333333')

sm = ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, ax=axes, orientation='vertical',
                    fraction=0.013, pad=0.015, aspect=30)
cbar.set_label(r'$\Delta E_\mathrm{d} = E_\mathrm{d,A} - E_\mathrm{d,B}$ (kJ mol$^{-1}$)',
               fontsize=11, labelpad=10)
cbar.ax.tick_params(labelsize=10)
cbar.outline.set_linewidth(1.2)

fig.suptitle('Robeson Plots — Colored by Enthalpic Diffusion Selectivity $\\Delta E_d$',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
